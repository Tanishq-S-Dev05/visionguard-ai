```python
import numpy as np
import pandas as pd
```

```python
# Create a 1D array of numbers from 0 to 9
arr = np.arange(10)

# Print: shape, size, dtype, number of dimensions
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Dtype:", arr.dtype)
print("Dimensions:", arr.ndim)
```

```python
# Create a 3x3 matrix of numbers 1 to 9
matrix = np.arange(1, 10).reshape(3, 3)

# Extract: first row, last column, center element (5)
print("First row:", matrix[0])
print("Last column:", matrix[:, -1])
print("Center:", matrix[1, 1])
```

```python
# Create two arrays: a = [1,2,3], b = [[1],[2],[3]]
a = np.array([1, 2, 3])
b = np.array([[1], [2], [3]])

# Add them (broadcasting) and print result
result = a + b
print(result)
print("Result shape:", result.shape)
```

```python
# Create array: [10, 25, 30, 45, 50, 65, 70, 85, 90, 100]
arr = np.array([10, 25, 30, 45, 50, 65, 70, 85, 90, 100])

# Select only elements > 50 and < 90
filtered = arr[(arr > 50) & (arr < 90)]
print(filtered)
```

```python
# Create two 2x2 matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiply (not element-wise) and print
C = np.dot(A, B)
print(C)
```

```python
# Create a DataFrame of 5 students
data = {
    'Name': ['Rahul', 'Priya', 'Amit', 'Sneha', 'Vikram'],
    'Age': [21, 22, 20, 23, 21],
    'Marks': [85, 92, 78, 95, 88],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai']
}
df = pd.DataFrame(data)
print(df)
```

```python
# From above df, select students with Marks > 85
top_students = df[df['Marks'] > 85]

# Sort by Age descending
sorted_df = top_students.sort_values('Age', ascending=False)
print(sorted_df)
```

```python
# Group by City and find average Marks per city
city_avg = df.groupby('City')['Marks'].mean()
print(city_avg)
```

```python
# Create array with NaN values
arr = np.array([1.0, 2.0, np.nan, 4.0, np.nan, 6.0])

# Find: count of NaN, mean ignoring NaN, fill NaN with 0
print("NaN count:", np.isnan(arr).sum())
print("Mean (ignore NaN):", np.nanmean(arr))
print("Filled:", np.nan_to_num(arr, nan=0))
```

```python
# Create a 5x5 grayscale image (random pixels 0-255)
image = np.random.randint(0, 256, size=(5, 5))

# Normalize to 0-1 range
normalized = image / 255.0

# Print: original, normalized, min, max
print("Original:\n", image)
print("Normalized:\n", normalized)
print("Min:", normalized.min(), "Max:", normalized.max())
```
